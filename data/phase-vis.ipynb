```python
import numpy as np
import imageio
import sys
import glob
import os
from tqdm import tqdm
import matplotlib.pyplot as plt
imageio.plugins.freeimage.download()
import scipy.stats as st
from mpl_toolkits.axes_grid1 import make_axes_locatable

def plot_phase_functions(cosines, label):
    fig, ax = plt.subplots(subplot_kw={'projection': 'polar'})

    bins = np.linspace(0, np.pi, 35)

    weights = 1.9/(np.sin(np.arccos(cosines[0]))*cosines[0].shape[0])
    p = ax.hist(np.arccos(cosines[0]), label="sampled phase function", weights=weights, bins=bins, color="orange", histtype="bar")
    ax.hist(-np.arccos(cosines[0]), weights=weights, bins=(-bins)[::-1], color="orange", histtype="bar")

    ax.set_title(label)
    ax.set_xlabel("$\omega_i . \omega_o$")
    ax.legend()
    ax.grid()
    ax.set_yticks([])
    ax.set_theta_zero_location("N")
    #ax.set_ylabel("$p(n_z)$")
    fig.tight_layout()
    return fig, ax

def lambert_phase(cosTheta):
    theta = np.arccos(cosTheta)
    return 2*(np.sin(theta) - theta*cosTheta) / (3 * np.pi * np.pi)

def uniform_phase(cosTheta):
    return np.ones_like(cosTheta)

with plt.style.context("seaborn-v0_8-dark"):
    params = {'text.usetex' : False,
            'font.size' : 7,
            'font.family' : 'Linux Libertine'}
    plt.rcParams.update(params)

    normal_files = dict()


    for file in tqdm(glob.glob("testing/phase/*/*/*/*/*.bin")):
        cosines = np.reshape(np.fromfile(file, dtype=np.float64), (-1,1))#.T

        end_ind = "deg"
        deg_idx = file.find(end_ind)
        deg_idx = deg_idx if deg_idx != -1 else -4

        key = file[:deg_idx+len(end_ind)] + ".bin"
        if key not in normal_files:
            normal_files[key] = []
        normal_files[key].append(cosines)



    common_prefix = os.path.commonprefix(list(normal_files.keys()))
    common_postfix = os.path.commonprefix([k[::-1] for k in list(normal_files.keys())])[::-1]
        
    for k, v in tqdm(normal_files.items()):
        if "beckmann" in k:
            continue

        print(k)

        normals = np.concatenate(v, axis=0)

        fig, ax = plot_phase_functions(v, k[len(common_prefix):])
        #plot_normal_collection(normals, k[len(common_prefix):-len(common_postfix)], k, bounds = ((-1, 1), (-1, 1)), vmin=0, vmax=2)
        #plot_normal_collection_1D(v, k[len(common_prefix):], k, gt_normals=normal_files[k.replace("-normals", "-normals-beckmann")])

        if "lambert" in k:
            mu = np.linspace(0, np.pi, 100)
            ax.plot(mu, lambert_phase(np.cos(mu)), color="blue", label="classical phase function")
            ax.plot(-mu, lambert_phase(np.cos(mu)), color="blue")
        elif "mirror" in k:
            mu = np.linspace(0, np.pi, 100)
            ax.plot(mu, uniform_phase(np.cos(mu)) / (4*np.pi), color="blue", label="classical phase function")
            ax.plot(-mu, uniform_phase(np.cos(mu)) / (4*np.pi), color="blue")

        ax.legend()

        fig.tight_layout()
        fig.show()

        ax.set_title("")
        fig.tight_layout()
        os.makedirs("print/" + k, exist_ok=True)

        print(k)
        fig.savefig("print/" + k+".pdf", bbox_inches='tight', pad_inches=0, dpi=300)

        plt.show()
```
