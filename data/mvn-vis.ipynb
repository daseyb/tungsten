```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal

mvn = multivariate_normal([1,2], [[1,0.5],[0.5,1]])

samples =  np.reshape(np.fromfile("testing/basic/mvn/samples.bin", dtype=np.float64), (-1,2))

gt_samples = mvn.rvs(10000)

plt.scatter(samples[:,0], samples[:,1], alpha=0.05)
plt.scatter(gt_samples[:,0], gt_samples[:,1], alpha=0.05)
plt.scatter(samples[:,0].mean(), samples[:,1].mean(), alpha=0.8)
plt.scatter(gt_samples[:,0].mean(), gt_samples[:,1].mean(), alpha=0.8)
plt.xlim(-3,6)
plt.ylim(-3,6)
plt.gca().set_aspect("equal")
plt.show()

pdf_ps = np.reshape(np.fromfile("testing/basic/mvn/pdf_points.bin", dtype=np.float64), (-1,2))
grid_size = int(np.sqrt(pdf_ps.shape[0]))
pdf_ps = pdf_ps.reshape(grid_size, grid_size,2)
pdf_values = np.reshape(np.fromfile("testing/basic/mvn/pdf.bin", dtype=np.float64), (grid_size,grid_size))
CS = plt.contourf(pdf_ps[...,0],pdf_ps[...,1],pdf_values)
cbar = plt.colorbar(CS)
plt.scatter(samples[:,0], samples[:,1], alpha=0.3, zorder=1, s=1)
plt.show()


CS = plt.contourf(pdf_ps[...,0],pdf_ps[...,1],mvn.pdf(pdf_ps))
cbar = plt.colorbar(CS)
plt.show()
```
